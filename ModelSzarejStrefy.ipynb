# Reading data

```python
from openpyxl import load_workbook
wb = load_workbook('śląskie.xlsx', read_only=True, keep_links=False)
sheetnames = wb.sheetnames
print(sheetnames)
```

```python
df_data = pd.read_excel('śląskie.xlsx', sheet_name="znormalizowane_R(t)", index_col=0)
#df_data = pd.read_excel('JS-Dane-podział-na-woj_pełne.xlsx', sheet_name="Kujawsko-pomorskie", index_col=0)
#df_data.set_index(df_data.columns[0])
#df_data=df_data[['x1', 'x2','x3', 'x4', 'x5','y1', 'y2', 'y3', 'y4', 'y5']]
df_data =df_data[['F1','F3','F4','F6','F7','F8','y']]
df_data
```

# MIMIC

```python
#do MIMIC
import semopy
# The first step
from semopy import ModelMeans
from semopy import Model
from semopy import ModelEffects

#model guaranteeing good fit for all voiderships
mod = """ eta1 ~ F1 + F3 + F4 + F6 + F7+ F8
eta1 =~   1*y """

#model = ModelEffects(mod)
#model = ModelMeans(mod)
model = Model(mod)

res = model.fit(df_data, obj= "DWLS")


stats = semopy.calc_stats(model)
#print(stats.T)

print("=============================")
print(mod)
print("CFI = ", stats.loc[:,"CFI"].values[0])
print("GFI = ", stats.loc[:,"GFI"].values[0])
print("AGFI = ", stats.loc[:,"AGFI"].values[0])
print("NFI = ", stats.loc[:,"NFI"].values[0])
print("RSMA = ", stats.loc[:,"RMSEA"].values[0])



factors = pd.DataFrame(model.predict_factors(df_data))
print(factors)
factors.to_excel("Eta_ślaskie_DWLS.xlsx")
```
